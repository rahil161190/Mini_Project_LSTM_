```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len= max([len(x) for x in input_sequences])
```

```python
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
x =  padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=283)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(283,100,input_length=max_len-1))
model.add(LSTM(150))
model.add(Dense(283,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.build(input_shape=(None, 56))
model.summary()
```

```python
model.fit(x,y,epochs=100)
```

```python
from IPython.core.displayhook import tokenize
text = 'mail'
#tokenize
for i in range(5):
  token_text = tokenizer.texts_to_sequences([text])[0]
  # here tokenizer is variable that was fitted on your faqs data earlier.
  #padding
  padded_tokenized_text = pad_sequences([token_text],maxlen=56,padding='pre')
  #predict
  pos = np.argmax(model.predict(padded_tokenized_text))

  for word,index in tokenizer.word_index.items():
    if pos == index:
      text =  text+ " " + word
      print(text)
```
